```python
from datetime import datetime, timedelta, date
import gzip
import glob
import csv
import pandas as pd
import os

def open_read_DMSP_SSJ(iyyyy,imm,idd,sat_num):
    '''
    SSJデータのファイルをオープンしてバイナリデータを読み込む
    
    ## inputパラメータ
        iyyyy:年(integer)
        imm:月(integer)
        idd:日(integer)
        sat_num:衛星番号（integer）
      
    ## outputパラメータ
        data:バイナリlistデータ
        SSJ_num:SSJの番号5か4(integer)
    '''
    ### set data path ###
    #HD内
    fpath="/Volumes/HD-PUSU3-D/DMSP/f16/"+ str(iyyyy).zfill(4) + '/' + str(imm).zfill(2) + '/'#'E:\\DMSP\\f16\\'
    #PC内
    #fpath='C:\\Users\\81907\\Documents\\f16\\'
    
    if iyyyy >= 2000:
        YY=iyyyy-2000
    else:
        YY=iyyyy-1900
        
    ### calculate day of year
    dt=datetime(iyyyy,imm,idd)
    sdate=datetime(iyyyy,1,1)
    doy_time=dt-sdate
    DOY=doy_time.days+1
    ###
    
    ### open gzfile ###
    source_file = fpath+'j5f'+str(sat_num)+str(YY).zfill(2)+str(DOY).zfill(3)+'.gz'
    target_file = fpath+'j5f'+str(sat_num)+str(YY).zfill(2)+str(DOY).zfill(3)
    
    with gzip.open(source_file, mode="rb") as gzip_file:
        content = gzip_file.read()
        with open(target_file, mode="wb") as decompressed_file:
            decompressed_file.write(content)

  
    ### search files ###
    fnames=glob.glob(fpath+'j?f'+str(sat_num)+str(YY).zfill(2)+str(DOY).zfill(3))
    if not fnames:
        print("No SSJ data from DMSP "+str(sat_num)+ ' on ' +str(iyyyy).zfill(4)+'-'+str(imm).zfill(2)+'-'+str(idd).zfill(2))
        return False
    else:
        print("Find file of "+str(fnames))
        # ssj_num=int(str(fnames)[-6])   コードミスであるため書き換え　正しくは[-11]
        ssj_num=int(str(fnames)[-11])
        

        #file open and read data
        if len(fnames) > 0 :
            for file in fnames:
                data=[]
                with open(file,'rb') as f: 
                    while True:
                        bytes=f.read(2)
                        if bytes:
                            data.append(int.from_bytes(bytes,byteorder='big'))
                        else:
                            break
                os.remove(file)
                print(file)
            return data,ssj_num
        else:
            return False


        

def read_DMSP_SSJ(iyyyy,imm,idd,sat_num):
    '''
    SSJデータを読み込むためのツール
          sub_read_DMSP_SSJ_pos,get_DMSP_gfactor,open_read_DMSP_SSJ
    と合わせて使用する
    ## inputパラメータ
        iyyyy:年(integer)
        imm:月(integer)
        idd:日(integer)
        sat_num:衛星番号（integer）
    ##　outputパラメータ
        pos_SSJ_data:SSJの位置が入っている　（毎分値）
        ie_data:イオン&電子フラックス（毎秒値が入っている）
    
    '''
    data, ssj_num=open_read_DMSP_SSJ(iyyyy,imm,idd,sat_num)
    
    ## set parameter　パラメータの設定
    d_ch_order=[4,3,2,1,8,7,6,5,12,11,\
                10,9,16,15,14,13,20,19,18,17]
    Ei=[30000,20400,13900,9450,6460,4400,3000,2040,1392,949,\
        949,646,440,300,204,139,95,65,44,30]
    Ei_new=[30000,20400,13900,9450,6460,4400,3000,2040,1392,\
        949,646,440,300,204,139,95,65,44,30]
    dEi=[9600,8050,5475,3720,2525,1730,1180,804,545.5,373,\
        373,254.5,173,118,80.5,54.5,37,25.5,17.5,14]
    dEi_new=[9600,8050,5475,3720,2525,1730,1180,804,545.5,\
        373,254.5,173,118,80.5,54.5,37,25.5,17.5,14]
    i_gf=get_DMSP_gfactor(sat_num)
    e_gf=get_DMSP_gfactor(sat_num,sp='electron')
    doy_f=datetime(iyyyy,1,1)

    if ssj_num == 5: ##SSJが5の場合dt=0.05秒を設定
        dt=0.05
    else:##SSJ45の場合dt=0.098秒を設定
        dt=0.098

    ## setparam
    dt_par=[]
    e_count,e_dflux,e_eflux,e_flag=[],[],[],[]
    i_count,i_dflux,i_eflux,i_flag=[],[],[],[]
    ##
    
    pos_SSJ_data=sub_read_DMSP_SSJ_pos(iyyyy,data)

    for im_index in range(0,len(data),2640): 
        doy_fd=datetime(data[im_index+4]-50+2000,1,1)
        if doy_fd==doy_f:
            for isec in range(0,60):
                sec_index=43*isec+im_index
        
                if data[im_index+1]== data[15+sec_index] and data[im_index+2]==data[16+sec_index] :
                    if data[2596-1+im_index] == 1:
                        secc=int((data[17+sec_index])/1000)
                        msec=data[17+sec_index] % 1000
                        dt_par.append(doy_fd+timedelta(days=data[im_index]-1,hours=data[15+sec_index], minutes=data[16+sec_index],\
                                   seconds=secc,microseconds=msec))
                    else:
                        secc=int(data[17+sec_index])
                        dt_par.append(doy_fd+timedelta(days=data[im_index]-1,hours=data[15+sec_index], minutes=data[16+sec_index],\
                                   seconds=secc))                
                    one_e_ji,one_e_jei,one_e_ci,one_e_flag=[np.nan]*19,[np.nan]*19,[np.nan]*19,np.nan
                    for ich in range(0,20):
                        if ich != 9 :
                            ida_ch=d_ch_order[ich]-1
                            I_ich=data[18+ich+sec_index]
                            if I_ich < 65536/2:
                                X_ch=(I_ich % 32)
                                Y_ch=(I_ich-X_ch)/32
                                Ci=(X_ch+32)*pow(2,Y_ch)-33
                                Ji=Ci/e_gf[ida_ch]/dt
                                JEi=Ji*Ei[ida_ch]
                            else:
                                Ci=np.nan
                                Ji=np.nan
                                JEi=np.nan
                            if ida_ch > 10:
                                ida_ch=ida_ch-1
                            one_e_ci[ida_ch]=Ci
                            one_e_ji[ida_ch]=Ji
                            one_e_jei[ida_ch]=JEi
                        else:
                            one_e_flag=I_ich
                            
                    one_i_ji,one_i_jei,one_i_ci,one_i_flag=[np.nan]*19,[np.nan]*19,[np.nan]*19,np.nan
                    for ich in range(0,20):
                        if ich != 9 :
                            ida_ch=d_ch_order[ich] -1
                            I_ich=data[38+ich+sec_index]
                            if I_ich < 65536/2:
                                X_ch=(I_ich % 32)
                                Y_ch=(I_ich-X_ch)/32
                                Ci=(X_ch+32)*pow(2,Y_ch)-33
                                Ji=Ci/i_gf[ida_ch]/dt
                                JEi=Ji*Ei[ida_ch]
                                
                                
                                # print("i_gf:",i_gf[ida_ch])
                                # print("Ci:",Ci)
                                # print("Ji:",Ji)
                                # print("JEi:",Ji)
                            else:
                                Ci=np.nan
                                Ji=np.nan
                                JEi=np.nan
                            if ida_ch > 10:
                                ida_ch=ida_ch-1
                            one_i_ci[ida_ch]=Ci
                            one_i_ji[ida_ch]=Ji
                            one_i_jei[ida_ch]=JEi
                        else:
                            one_i_flag=I_ich
                            
                    e_count.append(one_e_ci)
                    e_dflux.append(one_e_ji)
                    e_eflux.append(one_e_jei)
                    e_flag.append(one_e_flag)
                    i_count.append(one_i_ci)
                    i_dflux.append(one_i_ji)
                    i_eflux.append(one_i_jei)
                    i_flag.append(one_i_flag)

    
    ie_data={\
        'E_eV':Ei_new,\
        'dE_eV':dEi_new,\
        'part_time':dt_par,\
        'e_count':e_count,'e_dif_flux':e_dflux,'e_dif_eflux':e_eflux,'e_flag':e_flag,\
        'i_count':i_count,'i_dif_flux':i_dflux,'i_dif_eflux':i_eflux,'i_flag':i_flag}
                    
    return pos_SSJ_data,ie_data


def sub_read_DMSP_SSJ_pos(iyyyy,bdata):
    ## SSJデータのサブルーチン

    doy_f=datetime(iyyyy,1,1)
    pos_dt=[]
    pos_gdlat=[]
    pos_gdlon=[]
    pos_alt=[]
    pos_geolat_100=[]
    pos_geolon_100=[]
    pos_gmaglat_100=[]
    pos_gmaglon_100=[]
    pos_mlt=[]
    mile2meter=1852 # nautical miles(海里)をメートルに変換
    
    for im_index in range(0,len(bdata),2640):
        #d_year_arr.append[2000+dd[im_index+4]-50]
        #位置データの時間タグを定義する。位置データは1分値。
        doy_fd=datetime(bdata[im_index+4]-50+2000,1,1)
        if doy_fd==doy_f:
            pos_dt.append(doy_fd+timedelta(days=bdata[im_index]-1,hours=bdata[im_index+1],minutes=bdata[im_index+2],seconds=bdata[im_index+3]))
            
            #測地緯度の配列を作る
            if bdata[im_index+5] > 1800:
                pos_gdlat.append(float(bdata[im_index+5]-4995)/10.0)
            else:
                pos_gdlat.append(float(bdata[im_index+5]-900)/10.0)
            #測地経度の配列を作る
            pos_gdlon.append(float(bdata[im_index+6])/10.0)
            #測地高度の配列を作る
            pos_alt.append(float(bdata[im_index+7])*mile2meter)
            #地理緯度の配列を作る
            if bdata[im_index+8] > 1800:
                pos_geolat_100.append(float(bdata[im_index+8]-4995)/10.0)
            else:
                pos_geolat_100.append(float(bdata[im_index+8]-900)/10.0)
            #地理経度の配列を作る
            pos_geolon_100.append(float(bdata[im_index+9])/10.0)
            #地磁気緯度の配列を作る
            if bdata[im_index+10] > 1800:
                pos_gmaglat_100.append(float(bdata[im_index+10]-4995)/10.0)
            else:
                pos_gmaglat_100.append(float(bdata[im_index+10]-900)/10.0)
            #地磁気経度の配列を作る
            pos_gmaglon_100.append(float(bdata[im_index+11])/10.0)
            #地磁気地方時の配列を作る
            pos_mlt.append(float(bdata[im_index+12])+float(bdata[im_index+13])/60.+float(bdata[im_index+13])/3600.)
    pos_data={\
        'pos_time':pos_dt,\
        'pos_gdlat':pos_gdlat,\
        'pos_gdlon':pos_gdlon,\
        'pos_alt_meter':pos_alt,\
        'pos_geo_lat_100km':pos_geolat_100,\
        'pos_geo_lon_100km':pos_geolon_100,\
        'pos_gmag_lat_100km':pos_gmaglat_100,\
        'pos_gmag_lon_100km':pos_gmaglon_100,\
        'pos_mlt':pos_mlt}
    return pos_data
```
